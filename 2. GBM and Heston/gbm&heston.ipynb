```python
# Parametri plausibili
steps = 252        # passi giornalieri su 1 anno
T = 1.0            # 1 anno
paths = 10         # 10 percorsi simulati
r = 0.03           # tasso risk-free 3%
kappa = 2.0        # velocità mean-reversion
theta = 0.04       # long-term variance (4% = vol 20%)
sigma = 0.3        # vol of vol
rho = -0.7         # correlazione negativa asset-varianza
S0 = 100           # prezzo iniziale asset
v0 = 0.04          # varianza iniziale
seed = 42          # per riproducibilità

# Simulazione Heston
S_paths, v_paths = gh.generate_heston(
    steps=steps,
    T=T,
    paths=paths,
    r=r,
    kappa=kappa,
    theta=theta,
    sigma=sigma,
    rho=rho,
    S0=S0,
    v0=v0,
    seed=seed
)


gh.plot_simulation(S=S_paths, T=1, v=v_paths)
```
