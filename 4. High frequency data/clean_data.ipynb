```python
import pandas as pd
import os
from pathlib import Path
from tqdm import tqdm
```

```python
root_folder = r"D:\stockData\FTSE_MIB_stocks_merged"
output_folder = r"D:\stockData\FTSE_MIB_stocks_cleaned"
os.makedirs(output_folder, exist_ok=True)

root_folder = Path(root_folder)
stocks = [f.stem for f in root_folder.glob("*.parquet")]
```

```python
stocks = ['A2_MI', 'AMPF_MI'] ##  reduced list of stocks for testing
```

```python

for stock in tqdm(stocks):

    # Take one stock and save it in a pandas DataFrame
    path = os.path.join(root_folder, stock + '.parquet')
    df = pd.read_parquet(path)


    # Cleaning operations (one stock at the time)


    # Convert to date time and order
    df['DateTime'] = pd.to_datetime(df['DateTime'], errors='coerce')
    df = df.sort_values('DateTime') #.reset_index(drop=True)

    # Remove consecutive duplicates
    mask = df[stock] != df[stock].shift()
    df = df[mask]

    # Save cleaned stock in a parquet file stored in output_path folder
    output_path = os.path.join(output_folder, f"{stock}.parquet")
    df.to_parquet(output_path)
```
